```python
(
    tick
    .merge(prod)
    .groupby(["PRODUCTO","DESC_PRODUCTO"])
    .agg({
        "NUM_ITEMS": np.sum,#Cuantos items hay
        "TICKET": np.max # el ticket más actual en el que aparece 
    })
    .head(10)
)
```
